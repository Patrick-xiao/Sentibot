# Import some necessary library

```python
!pip install evaluate
!pip install datasets
import torch
import pandas as pd
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from sklearn.model_selection import train_test_split
from datasets import Dataset
```

# Import our sentiment dataset

```python
from google.colab import drive
drive.mount('/content/drive/')
data_path = "/content/drive/MyDrive/CDS525_Deep_Learning/Bert/"  # this is your drive
df = pd.read_csv(data_path+"sentiment_analysis.csv")
new_df = df[["text","sentiment"]]

# 将情感标签映射为数字
label_mapping = {"positive": 0, "neutral": 1, "negative": 2}
new_df["label"] = new_df["sentiment"].map(label_mapping)

# 划分训练集和验证集
train_df, val_df = train_test_split(new_df, test_size=0.2, random_state=42)
```

# Use tokenizer to process the text

```python
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True, padding=True)
```

# Train the distillation model

```python
train_dataset = Dataset.from_pandas(train_df).map(preprocess_function, batched=True)
val_dataset = Dataset.from_pandas(val_df).map(preprocess_function, batched=True)
train_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])
val_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])
```

```python
model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=3)
# freeze base model parameters
for name, param in model.base_model.named_parameters():
    param.requires_grad = True

# # unfreeze base model pooling layers
# for name, param in model.base_model.named_parameters():
#     if "attention" in name:
#         param.requires_grad = True
# # print layers
# for name, param in model.named_parameters():
#    print(name, param.requires_grad)
```

```python
# 定义训练参数
training_args = TrainingArguments(
    output_dir=data_path,
    num_train_epochs=5,  # 训练轮数
    per_device_train_batch_size=8,  # 训练批次大小
    per_device_eval_batch_size=8,  # 评估批次大小
    warmup_steps=50,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    report_to="none"  # 新增这一行，不使用 wandb 记录
)
```

```python
# 计算准确率的函数
import evaluate

# 计算准确率的函数
accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return accuracy.compute(predictions=predictions, references=labels)

# 训练模型
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)

trainer.train()
```

# Apply the model

```python
from transformers import AutoModelForSequenceClassification

from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch

# 定义保存模型的目录
output_dir ='/content/drive/MyDrive/CDS525_Deep_Learning/Bert/checkpoint-250/'  # 根据实际情况修改

# 加载分词器
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')

# 加载微调后的模型
model = AutoModelForSequenceClassification.from_pretrained(output_dir)

# 输入文本
text = "Hey, I don't like it!"

# 对输入文本进行分词和编码
inputs = tokenizer(text, return_tensors='pt')

# 使用模型进行预测
with torch.no_grad():
    outputs = model(**inputs)

# 获取预测结果
logits = outputs.logits
predicted_class_id = logits.argmax().item()

# 映射预测结果到标签
id2label = model.config.id2label
predicted_label = id2label[predicted_class_id]

print(f"Predicted label: {predicted_label}")
```

```python
# Apply model on test dataset
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))
```

```python
val_df['sentiment_score'] = val_df['text'].apply(lambda x: sentiment_score(x[:]))
val_df.head(100)
```

```python
def judge_sentiment(score):
  if score==0:
    return "positive"
  elif score==1:
    return "neutral"
  else:
    return"negative"
val_df["predict_sentiment"] = val_df["sentiment_score"].apply(lambda x: judge_sentiment(x))
# Put the predicted result and their raw labels into a csv file
final_df = val_df.drop(columns=["sentiment_score"])
final_df.to_csv('/content/drive/MyDrive/CDS525_Deep_Learning/Bert/performance_evaluation4.csv', index=False)
```

```python
correct_prediction = (final_df["sentiment"]==final_df["predict_sentiment"]).sum()
total_num = len(final_df)
accuracy = correct_prediction/total_num
print(f"accuracy:{accuracy*100:.2f}%")
```
